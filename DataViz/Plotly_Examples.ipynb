Importando o plotly e habilitando o modo offline:

```python
from plotly.offline import iplot
import plotly.graph_objs as go
```

Base de dados pública utilizada: Informações de vendas a partir de 2017 da Olist, um E-commerce brasileiro ([Para saber mais](https://www.kaggle.com/olistbr/brazilian-ecommerce/version/2)).

```python
import pandas as pd

urlOlistVend = 'https://raw.githubusercontent.com/brunamulinari/DataBasesPython/master/Olist/olist_classified_public_dataset.csv'
dfOlistV = pd.read_csv(urlOlistVend)
dfOlistV.head(n=5)
```

Exibindo informações sobre o meu banco de dados:

```python
dfOlistV.info()
```

Nessa seção vamos criar visualizações para responder algumas perguntas sobre a base de dados:

#### Qual a distribuição dos status dos pedidos?

```python
quantStatus = dfOlistV.order_status.value_counts()
quantStatus
```

```python
data = [go.Bar(x = quantStatus.index,
                    y = quantStatus
                  )]

layout = go.Layout(title={
                  'text': 'Status dos pedidos',
                  'y':0.9,
                  'x':0.5,
                  'xanchor': 'center',
                  'yanchor': 'top'},
                   xaxis = {'title': 'Status'},
                   yaxis = {'title': 'Quantidade'},
                   barmode = 'stack')

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

#### Qual o mês do ano que houve mais vendas?

```python
# Transformando a coluna em datetime
dfOlistV.order_purchase_timestamp = pd.to_datetime(dfOlistV.order_purchase_timestamp)
# Obter o formato em meses
meses = dfOlistV.order_purchase_timestamp.dt.to_period('M')
# Armazenando em uma nova coluna com o formato string
dfOlistV['order_purchase_month'] = meses.astype(str)
dfOlistV.order_purchase_month.head()
```

```python
dfOlistV['order_purchase_month']
```

```python
vendasPMes = dfOlistV.groupby('order_purchase_month')['order_products_value'].sum()
vendasPMes
```

```python
trace1 = go.Scatter(x=vendasPMes.index, 
                   y=vendasPMes,
                   name = '',
                   mode='lines')

trace2 = go.Bar(x=vendasPMes.index, 
               y=vendasPMes,
               name = '',
               opacity = 0.7,
               marker = {'color':'lightskyblue'})

layout = go.Layout(title={
                  'text': 'Vendas no ano',
                  'y':0.9,
                  'x':0.5,
                  'xanchor': 'center',
                  'yanchor': 'top'},
                   xaxis = {'title': 'Período'},
                   yaxis = {'title': 'Valor de vendas'})

data = [trace1, trace2]

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

Destacando uma barra

```python
# Valor máximo de vendas
maxVen = vendasPMes.values.max()

cores = []
for i,n in zip(vendasPMes.values, vendasPMes.index):
  if (i == maxVen):
    mesMax = n
    cores.append('blue')
  else:
    cores.append('lightskyblue')
```

```python
trace = go.Bar(x=vendasPMes.index, 
               y=vendasPMes,
               name = '',
               opacity = 0.7,
               marker = {'color':cores,
                         'line':{'color': '#333',
                                 'width': 2}})

layout = go.Layout(title={
                  'text': 'Vendas no ano',
                  'y':0.9,
                  'x':0.5,
                  'xanchor': 'center',
                  'yanchor': 'top'},
                   xaxis = {'title': 'Período'},
                   yaxis = {'title': 'Valor de vendas'},
                   annotations=[{'text': 'Mês destaque',
                                 'x': mesMax,
                                 'y': maxVen}])

data = [trace]

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

#### Qual foi a quantidade de vendas por categoria?

```python
# Contabiliza os produtos por categoria
prodPorCat = dfOlistV.groupby('product_category_name').id.count()

# Filtra categorias com mais de 200 vendas e ordena do maior para o menor
prodPorCat = prodPorCat.loc[prodPorCat > 200]
prodPorCat.sort_values(ascending=False,inplace=True)
print(prodPorCat)
```

```python
trace = go.Pie(labels=prodPorCat.index,
               values=prodPorCat,
               direction='clockwise',
               marker = {'colors':prodPorCat,
                         'line':{'color': '#000000', 'width':1}},
               pull=[0.1,0,0,0,0])

layout = go.Layout(title={
                  'text': "Quantidade de Vendas por Categoria",
                  'y':0.9,
                  'x':0.35,
                  'xanchor': 'center',
                  'yanchor': 'top'},
                   font={
                  'family':"Courier New, monospace",
                  'size': 18,
                  'color':"RebeccaPurple"}
 )

data = [trace]

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

#### O valor do frete aumenta com o aumento do preço do produto?

```python
trace = go.Scatter(x = dfOlistV['order_freight_value'],
                  y = dfOlistV['order_products_value'],                
                  mode='markers',
                  text = 'Status do pedido: '+ dfOlistV['order_status'] + '<br>'+
                  'Classe: ' + dfOlistV['most_voted_class'],
                  hoverinfo='text+x+y',
                  marker = {'color':'forestgreen'})

data = [trace]

layout = go.Layout(title={
                  'text': "Valor do frete x Valor do produto",
                  'y':0.9,
                  'x':0.5,
                  'xanchor': 'center',
                  'yanchor': 'top'},
                   xaxis = {'title': 'Valor do Frete',
                            'tickformat':'.',
                            'tickprefix':'R$ '},
                   yaxis = {'title': 'Valor do Produto',
                            'tickformat':'.',
                            'tickprefix':'R$ '}
 )

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```
